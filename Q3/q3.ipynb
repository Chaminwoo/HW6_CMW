```python
import csv
import matplotlib.pyplot as plt

def read_csv(filename):
    with open(filename, 'r', encoding='ANSI') as f:
        data = list(csv.reader(f))
    return data[1:] 
def calculate_population(data):
    total_male = 0
    total_female = 0
    for row in data:
        if '(5000000000)' in row[0]:
            total_male += int(row[1].replace(',', ''))
            total_female += int(row[104].replace(',', ''))
    return total_male/10000, total_female/10000

def plot_barh_and_pie(title, total_male, total_female, colors):
    labels = ['Male', 'Female']
    sizes = [total_male, total_female]

    fig, ax = plt.subplots(1, 2, figsize=(10, 5))

    ax[0].set_title(title)
    ax[0].barh('Female', total_female, color=colors[1], label="Female")
    ax[0].barh('Male', total_male, color=colors[0], label="Male")
    ax[0].set_xlabel('Population (Unit: 10000)')
    ax[0].legend()
    ax[0].text(total_female, 'Female', f'{total_female:,}', va='center')
    ax[0].text(total_male, 'Male', f'{total_male:,}', va='center')

    ax[1].set_title(title)
    ax[1].pie(sizes, labels=labels, colors=colors, autopct='%1.3f%%')
    plt.show()

def main():
    filenames = ['q3.csv', 'q3(1).csv', 'q3(2).csv']
    colors = [['cyan', 'orange'], ['yellow', 'pink'], ['blue', 'purple']]
    titles = ['Male and Female Populations in 2022',
              'Male and Female Populations in 2021',
              'Male and Female Populations in 2020']

    for filename, color, title in zip(filenames, colors, titles):
        data = read_csv(filename)
        total_male, total_female = calculate_population(data)
        plot_barh_and_pie(title, total_male, total_female, color)

if __name__ == "__main__":
    main()
```
